```python
# zad L6.3
from scipy.integrate import quad, trapz, simps
import numpy as np

def f(x):
    return x**4 + 12*x**3 + 3*x**2 + 6*x

n_values = [3, 5, 9, 17, 33, 49, 65, 81, 97, 113, 129, 145, 161, 177, 193, 209, 225, 241, 257, 273, 289]

for n in n_values:
    dokl, error = quad(f, 0, 5)
    x = np.linspace(0, 5, n)
    trapez = trapz(f(x), x)
    err_trapz = abs(dokl-trapez)
    print(f"n = {n:3d}, wartość = {trapez:.5f}, błąd = {err_trapz:.5g}")
    
print("-----")
for n in n_values:
    dokl, error = quad(f, 0, 5)
    x = np.linspace(0, 5, n)
    simpson = simps(f(x), x)
    err_simpson = abs(dokl-simpson)
    print(f"n = {n:3d}, wartość = {simpson:.5f}, błąd = {err_simpson:.5g}")
```
