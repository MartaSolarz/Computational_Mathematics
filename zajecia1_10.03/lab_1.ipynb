## Zadanie 1

Wykonaj funkcję, która zwraca sumę oraz różnicę dwóch wektorów, podanych jako parametry.

```python
import numpy as np

def calculate(w,v):
    return w+v, w-v
```

## Zadanie 4

Napisz funkcję, której parametrami są funkcja rzeczywista $f$, brzegi przedziału $[a, b]$  oraz liczba naturalna $n$, która przybliża wartość
$ max\left(f(x): x \in [a,b]\right) \approx max\left(f(x_i): i \in \{1,2,..., n\}\right)$
gdzie $x_i$ są równoodległymi węzłami w przedziale $[a, b]$.

```python
def approx_max(f, a, b, n):
    # Równoodległe węzły na przedziale [a,b]
    x = np.linspace(a, b, n)

    # Obliczenie wartości funkcji f w węzłach
    f_values = f(x)

    # Zwrócenie największej wartości z f_values
    return max(f_values)
```

```python
def f(x):
    return x**2 + 2*x + 1

approx = approx_max(f, -1, 1, 1000)
print(f"Przybliżone maksimum: {approx:.5f}")
```

## Zadanie 5

Znajdź elementy $x_0, . . . , x_{40}$ ciągu określonego zależnością:
$x_n = \frac{13}{3}x_{n−1} − \frac{4}{3}x_{x-2}$
dla warunków początkowych:
- $x_0 = 1, x_1 = 1$
- $x_0 = 1, x1 = 4$

Określ błąd względny obliczenia każdego z elementów $x_0, . . . , x_{40}$ w arytmetyce zmiennoprzecinkowej względem wartości otrzymanej w arytmetyce dokładnej.

```python
def compute_sequence(start1, start2, n):
    sequence = [start1, start2]

    for i in range(2, n+1):
        x_i = (13/3) * sequence[i-1] - (4/3) * sequence[i-2]
        sequence.append(x_i)

    return sequence
```

```python
import sympy as sp

sequence_A = compute_sequence(1, 1, 40)

for i, x in enumerate(sequence_A):
    exact = sp.Rational(3**i - 1, 2)
    error = abs(x - exact) / exact

    print(f"x_{i}: {x:.2f}, relative error: {error}")
```

```python
sequence_B = compute_sequence(1, 4, 40)

for i, x in enumerate(sequence_B):
    exact = sp.Rational(3**i - 1, 2)
    error = abs(x - exact) / exact

    print(f"x_{i}: {x:.2f}, relative error: {error}")
```

## Zadanie 6

Wykonaj wykresy wielomianu
$w(x) := {(x − 1)}^4 = x^4 − 4x^3 + 6x^2 − 4x + 1$
w węzłach równoodległych na przedziale $[0.9995, 1.0005]$ korzystając z obydwu tych wzorów.
Który wzór jest dokładniejszy i dlaczego?

```python
import numpy as np
x = np.linspace(0.9995, 1.0005, 1000)
y1 = (x-1)**4
y2 = x**4 - 4*x**3 + 6*x**2 - 4*x + 1

plt.plot(x,y1, label="$w(x) := {(x − 1)}^4$")
plt.plot(x, y2, label="$v(x) := x^4 − 4x^3 + 6x^2 − 4x + 1$")
plt.grid()
plt.legend();
```

Dokładniejszy jest ten drugi. Z wykresu widać, że różnice ilorazowe między kolejnymi węzłami są bardzo bliskie zera, co sugeruje, że wzór na wartość wielomianu jest dokładniejszy. Ponadto, wzór na wartość wielomianu jest prostszy i mniej obciążony numerycznie niż wzór na różnicę ilorazową, co również przemawia za jego dokładnością.
